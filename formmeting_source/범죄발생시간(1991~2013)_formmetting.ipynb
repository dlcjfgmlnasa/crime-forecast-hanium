```python
import pandas as pd
import os
```

```python
포맷안된디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄발생 시간[1964-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된디렉토리,파일이름))
데이터프레임
```

```python
범죄범주리스트=[데이터 for 데이터 in 데이터프레임['범죄명칭'].values.tolist()]
데이터=''
for i,범주 in enumerate(범죄범주리스트) :
    if pd.notnull(범주) :
        데이터=범주
    범죄범주리스트[i]=데이터
범죄범주리스트=범죄범주리스트[3:]
데이터프레임=데이터프레임.drop('범죄명칭',axis=1)
데이터프레임
```

```python
범죄이름리스트=[]
for index in 데이터프레임.index :
    Temp=''
    for column in ['Unnamed: 1','Unnamed: 2','Unnamed: 3'] :
        word=데이터프레임.ix[index][column]
        if pd.notnull(word) :
            Temp+=word+','
    범죄이름리스트.append(Temp[:-1])

for i in range(3) : 범죄이름리스트[i]=i
```

```python
데이터프레임=데이터프레임.set_index([범죄이름리스트],drop=True)
데이터프레임=데이터프레임.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
데이터프레임
```

```python
데이터프레임리스트={str(year):[] for year in range(1991,2014)}
year=''
for 콜론 in 데이터프레임 :
    for 데이터 in range(1991,2014) :
        if 콜론 == str(데이터) : 
            year=콜론
            break
    데이터프레임리스트[year].append(데이터프레임[콜론])
```

```python
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임리스트[년도]=pd.concat(데이터프레임,axis=1)
```

```python
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=데이터프레임.drop([0,2])
    데이터프레임.columns = 데이터프레임.ix[0].values
    데이터프레임=데이터프레임.drop(1)
    데이터프레임.insert(0,column='범죄범주',value=범죄범주리스트)
    데이터프레임=데이터프레임.T
    데이터프레임리스트[년도]=데이터프레임
```

```python
writer=pd.ExcelWriter(os.path.join(타겟디렉토리,파일이름))
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임.to_excel(writer,년도)
writer.save()
```

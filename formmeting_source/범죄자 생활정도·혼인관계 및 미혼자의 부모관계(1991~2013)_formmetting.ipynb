```python
import pandas as pd
import os
```

```python
포맷안된디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄자 생활정도·혼인관계 및 미혼자의 부모관계[1972-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된디렉토리,파일이름))
데이터프레임
```

```python
범죄대분류리스트=[범죄 for 범죄 in 데이터프레임['범죄명칭']]
temp=''
for i, 범죄 in enumerate(범죄대분류리스트) :
    if pd.notnull(범죄) :
        temp=범죄
    범죄대분류리스트[i]=temp
범죄대분류리스트=list(filter(lambda x : x != '', 범죄대분류리스트))
범죄대분류리스트
```

```python
범죄소분류리스트=[범죄 for 범죄 in 데이터프레임['Unnamed: 1']]
temp=''
for i, 범죄 in enumerate(범죄소분류리스트) :
    if pd.notnull(범죄) :
        temp=범죄
    범죄소분류리스트[i]=temp
범죄소분류리스트=list(filter(lambda x : x!='', 범죄소분류리스트))
범죄소분류리스트
```

```python
인덱스리스트=[]
for index in 데이터프레임.index :
    word=''
    for column in ['Unnamed: 2','Unnamed: 3'] :
        temp=데이터프레임.ix[index][column]
        if pd.notnull(temp) : word+=temp+','
    if word=='' : word=데이터프레임.ix[index]['Unnamed: 1']
    else : word=word[:-1]
    인덱스리스트.append(word)
인덱스리스트=list(filter(lambda x : pd.notnull(x), 인덱스리스트))
인덱스리스트
```

```python
데이터프레임=데이터프레임.drop(['범죄명칭','Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
데이터프레임
```

```python
데이터프레임리스트={year:[] for year in range(1991,2014)}
for 콜론 in 데이터프레임 :
    try : temp=int(콜론)
    except : pass
    데이터프레임리스트[temp].append(데이터프레임[콜론])
```

```python
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(데이터프레임,axis=1)
    데이터프레임.columns = [i for i in range(20)]
    데이터프레임=데이터프레임.drop(4,axis=1)
    콜론리스트=[콜론[5:] for 콜론 in 데이터프레임.ix[3].values.tolist()]
    분류리스트=[]
    for i in range(len(콜론리스트)) :
        if i in range(0,4) : 분류리스트.append('생활정도')
        elif i in range(4,9) : 분류리스트.append('혼인관계')
        elif i in range(9,19) : 분류리스트.append('미혼자부모관계')
    분류리스트.insert(0,'')
    분류리스트.insert(1,'')
    데이터프레임=데이터프레임.drop([0,1,2,3])
    데이터프레임.columns = 콜론리스트
    데이터프레임=데이터프레임.set_index([인덱스리스트],drop=True)
    데이터프레임.insert(0,value=범죄대분류리스트,column='범죄대분류')
    데이터프레임.insert(1,value=범죄소분류리스트,column='범죄소분류')
    데이터프레임=데이터프레임.T
    데이터프레임.insert(0,value=분류리스트,column='생활정도/혼인관게/미혼자부모관게')
    데이터프레임=데이터프레임.T
    데이터프레임리스트[년도]=데이터프레임
```

```python
writer=pd.ExcelWriter(os.path.join(타겟디렉토리,파일이름))
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임.to_excel(writer,str(년도))
writer.save()
```

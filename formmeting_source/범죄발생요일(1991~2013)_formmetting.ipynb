```python
import pandas as pd
import os
```

```python
파일디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄발생 요일[1964-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(파일디렉토리,파일이름))
데이터프레임
```

```python
범죄분류리스트=[column for column in 데이터프레임['범죄명칭'].values.tolist()]
Temp=''
for i,범죄 in enumerate(범죄분류리스트):
    if not pd.isnull(범죄) : Temp=범죄
    범죄분류리스트[i]=Temp
print(범죄분류리스트)


count=0
for i in 데이터프레임.index :
    데이터프레임.ix[i]['범죄명칭']=범죄분류리스트[count]
    count=count+1
데이터프레임
```

```python
범죄명칭리스트=[]
for 인덱스 in 데이터프레임.index :
    Temp=''
    for 콜론 in ['Unnamed: 1','Unnamed: 2','Unnamed: 3'] :
        if pd.notnull(데이터프레임.ix[인덱스][콜론]) :
            Temp+=(데이터프레임.ix[인덱스][콜론])+','
    범죄명칭리스트.append(Temp[:-1])

데이터프레임=데이터프레임.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
```

```python
데이터프레임=데이터프레임.set_index([범죄명칭리스트])
데이터프레임
```

```python
년도=""
년도리스트=[]
count=0
for 인덱스 in 데이터프레임.columns[1:] :
    for year in range(1991,2014) :
        if 인덱스 == str(year) : 
            년도=인덱스
            count=0
            pass
    년도리스트.append(년도+'.'+str(count))
    count=int(count)+1
년도리스트.insert(0,'범죄종류')
데이터프레임.columns=년도리스트
데이터프레임
```

```python
범죄종류인덱스=데이터프레임['범죄종류']
데이터프레임=데이터프레임.drop('범죄종류',axis=1)
데이터프레임
```

```python
데이터프레임리스트={str(year) :[] for year in range(1991,2014)}
for 인덱스 in 데이터프레임 :
    index=인덱스.split('.')[0]
    Series=데이터프레임[인덱스]
    데이터프레임리스트[index].append(Series)
```

```python
최종리스트=[]
for 년도, each데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(each데이터프레임,axis=1)
    데이터프레임.insert(0,value=범죄분류리스트,column='범죄분류')
    데이터프레임=데이터프레임.T
    Temp=데이터프레임.columns.tolist()
    for i in range(3) : Temp[i]=i
    데이터프레임.columns=Temp
    Series=데이터프레임[1]
    요일리스트=Series.values
    요일리스트[0]='범죄분류'
    데이터프레임=데이터프레임.set_index([요일리스트],drop=True)
    데이터프레임=데이터프레임.drop([0,1,2],axis=1)
    데이터프레임=데이터프레임.T
    최종리스트.append((년도,데이터프레임))
```

```python
writer=pd.ExcelWriter(os.path.join(타겟디렉토리,파일이름))
for 년도, 데이터프레임 in 최종리스트 :
    데이터프레임.to_excel(writer,년도)
writer.save()
```

```python
import pandas as pd
import os
```

```python
포맷안된경로="C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일"
타겟경로="C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일"
파일이름="범행시 일기[1964-2013].xlsx"
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된경로,파일이름),encoding='utf-8')
데이터프레임
```

```python
범죄종류리스트=[]
상자=''
for 범죄 in 데이터프레임['범죄명칭'].values :
    if not pd.isnull(범죄) :
        상자=범죄
    범죄종류리스트.append(상자)
범죄종류리스트
```

```python
범죄명칭리스트=[]
for i in 데이터프레임.index :
    범죄명칭=""
    for 콜론 in ['Unnamed: 1','Unnamed: 2','Unnamed: 3'] :
        if not pd.isnull(데이터프레임[콜론].ix[i]) :
            범죄명칭+=데이터프레임[콜론].ix[i]+','
    범죄명칭=범죄명칭[:-1]
    범죄명칭리스트. append(범죄명칭)
범죄명칭리스트        
```

```python
데이터프레임=데이터프레임.set_index([범죄명칭리스트],drop=True)
데이터프레임=데이터프레임.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
데이터프레임
```

```python
count=0
for 열 in 데이터프레임['범죄명칭'].values :
    데이터프레임['범죄명칭'][count]=범죄종류리스트[count]
    count=count+1
데이터프레임
```

```python
년도리스트=[]
Temp=''
count=0
for 년도 in 데이터프레임.columns[1:] :
    for 년 in range(1991,2014) :
        if 년도==str(년) : 
            Temp=년도
            count=0
            pass
    년도리스트.append(Temp+'.'+str(count))
    count=(int)(count)+1

년도리스트.insert(0,'범죄명칭')
데이터프레임.columns=년도리스트
년도리스트
```

```python
리스트={년도:[] for 년도 in range(1991,2014)}

데이터프레임=데이터프레임.T
데이터프레임
```

```python
데이터프레임리스트={년도:[] for 년도 in range(1991,2014)}

for 인덱스 in 데이터프레임.index[1:] :
    for 년도 in 리스트.keys() :
        if int(인덱스.split('.')[0]) == 년도 :
            Temp=데이터프레임.ix[인덱스]
            데이터프레임리스트[년도].append(Temp)
            break
범죄종류리스트=범죄종류리스트[3:]
```

```python
상자=[]

for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    df=pd.concat(데이터프레임,axis=1)
    인덱스리스트=df.index.values
    인덱스리스트[0]=0
    인덱스리스트[1]=1
    인덱스리스트[2]=2
    df.set_index(인덱스리스트,drop=True)
    df.columns=df.ix[1].values
    df=df.drop([0,1,2])
    df.insert(0,value=범죄종류리스트,column='범죄종류')
    상자.append((년도,df))
상자[0][1]
```

```python
writer=pd.ExcelWriter(os.path.join(타겟경로,파일이름))
for 년도, 데이터프레임 in 상자 :
    데이터프레임.to_excel(writer,str(년도))
writer.save()
```

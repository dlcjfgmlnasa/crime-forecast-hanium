```python
import pandas as pd
import os
```

```python
포맷안된파일='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄자 범행시 연령[1964-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된파일,파일이름))
데이터프레임
```

```python
대분류리스트=데이터프레임['범죄명칭'].values.tolist()
word=''
for i, 범죄 in enumerate(대분류리스트) :
    if pd.notnull(범죄) :
        word=범죄
    대분류리스트[i]=word
대분류리스트=[범죄 for 범죄 in 대분류리스트 if 범죄 != '']
대분류리스트
```

```python
소분류리스트=데이터프레임['Unnamed: 1'].values.tolist()
word=''
for i, 범죄 in enumerate(소분류리스트) :
    if pd.notnull(범죄) :
        word=범죄
    소분류리스트[i]=word
소분류리스트=[범죄 for 범죄 in 소분류리스트 if 범죄 != '']
소분류리스트
```

```python
범죄이름리스트=[]
for index in 데이터프레임.index :
    word=''
    for column in ['Unnamed: 2','Unnamed: 3'] :
        범죄=데이터프레임.ix[index][column]
        if pd.notnull(범죄) :
            word+=범죄+','
    if word=='' : word=데이터프레임.ix[index]['Unnamed: 1']
    else : word=word[:-1]
    범죄이름리스트.append(word)
for i in range(4) : 범죄이름리스트[i]=i
데이터프레임=데이터프레임.set_index([범죄이름리스트])
데이터프레임=데이터프레임.drop(['범죄명칭','Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
데이터프레임
```

```python
데이터프레임리스트={year:[] for year in range(1991,2014)}
year=0
for 콜론 in 데이터프레임 :
    try : year=int(콜론)
    except : pass
    데이터프레임리스트[year].append(데이터프레임[콜론])
```

```python
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(데이터프레임,axis=1)
    #인구분류 리스트
    인구분류=데이터프레임.ix[1].values.tolist()
    Temp=''
    for i, 인구 in enumerate(인구분류) :
        if pd.notnull(인구) : Temp=인구
        인구분류[i]=Temp[:2]
        
    #나이 리스트
    나이리스트=데이터프레임.ix[2].values.tolist()
    for i, 나이 in enumerate(나이리스트) :
        if pd.isnull(나이) : 
            나이리스트[i]=데이터프레임.ix[1][i]
    데이터프레임.columns = 나이리스트
    데이터프레임=데이터프레임.drop([0,1,2,3])
    데이터프레임.insert(0,column='범죄대분류',value=대분류리스트)
    데이터프레임.insert(1,column='범죄소분류',value=소분류리스트)
    데이터프레임=데이터프레임.T
    인구분류.insert(0,'범죄분류')
    인구분류.insert(1,'범죄분류')
    데이터프레임.insert(0,column='인구분류',value=인구분류)
    데이터프레임리스트[년도]=데이터프레임
```

```python
writer=pd.ExcelWriter(os.path.join(타겟디렉토리,파일이름))
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임.to_excel(writer,str(년도))
writer.save()
```

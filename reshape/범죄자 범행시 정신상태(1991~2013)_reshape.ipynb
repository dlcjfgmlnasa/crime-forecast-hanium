```python
포맷안된디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄자 범행시 정신상태[1972-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된디렉토리,파일이름))
데이터프레임
```

```python
def makelist(columnName) :
    tempList=[column for column in 데이터프레임[columnName]]
    word=''
    for i, 범죄 in enumerate(tempList) :
        if pd.notnull(범죄) : word=범죄
        tempList[i]=word
        
    tempList=list(filter(lambda x : x!= '', tempList))
    return tempList
```

```python
대분류리스트=makelist('범죄명칭')
소분류리스트=makelist('Unnamed: 1')
소분류리스트
```

```python
Temp
범죄명리스트=[]
for index in 데이터프레임.index :
    word=''
    for column in ['Unnamed: 1','Unnamed: 2'] :
        Temp=데이터프레임.ix[index][column]
        if pd.notnull(Temp) : word+=Temp+','
    범죄명리스트.append(word[:-1])
범죄명리스트=list(filter(lambda x : x != '', 범죄명리스트))
범죄명리스트
```

```python
dropList=['범죄명칭','Unnamed: 1','Unnamed: 2','Unnamed: 3']
데이터프레임=데이터프레임.drop(dropList,axis=1)
데이터프레임
```

```python
데이터프레임리스트={year:[] for year in range(1991,2014)}
year=''
for column in 데이터프레임 :
    try : year=int(column)
    except : pass
    word=데이터프레임[column]
    데이터프레임리스트[year].append(word)
데이터프레임리스트[1991]
```

```python
Temp

for 년도,데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(데이터프레임,axis=1)
    콜론리스트=데이터프레임.ix[2].values.tolist()
    for i,분류 in enumerate(콜론리스트) :
        if pd.isnull(분류) : 콜론리스트[i]=데이터프레임.ix[1][i]
    정신상태리스트=데이터프레임.ix[1].values.tolist()
    
    데이터프레임.columns = 콜론리스트
    데이터프레임=데이터프레임.drop([0,1,2,3])
    데이터프레임=데이터프레임.set_index([범죄명리스트],drop=True)
    데이터프레임리스트[년도]=데이터프레임
```

```python
for 년도,데이터프레임 in 데이터프레임리스트.items() : 
    데이터프레임.insert(0,value=대분류리스트,column='범죄대분류')
    데이터프레임.insert(1,value=소분류리스트,column='범죄소분류')
    데이터프레임리스트[년도]=데이터프레임
데이터프레임리스트[1991]
```

```python
writer=pd.ExcelWriter(os.path.join(타겟디렉토리,파일이름))
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임.to_excel(writer,str(년도))
writer.save()
```

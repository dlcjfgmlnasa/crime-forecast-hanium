```python
연도리스트=[]
for 년도 in 데이터프레임.index[1:] :
    if 년도 in ['2011','2012','2013','2014'] :
        count=0
        인덱스=년도
    else : count=count+1
    연도리스트.append(인덱스+'.'+str(count))
연도리스트.insert(0,'직함')
연도리스트
```

```python
데이터프레임=데이터프레임.set_index([연도리스트],drop=True)
데이터프레임리스트={
    "2011":[],
    "2012":[],
    "2013":[],
    "2014":[],
}

for 인덱스 in 데이터프레임.index[1:] :
    데이터=데이터프레임.ix[인덱스]
    for 년도 in 데이터프레임리스트.keys() :
        if 인덱스.startswith(년도) :
            데이터프레임리스트[년도].append(데이터)
            pass
```

```python
최종리스트=[]
for 년도, 개별데이터프레임 in 데이터프레임리스트.items() :
    프레임=pd.concat(개별데이터프레임,axis=1)
    프레임.insert(0,value=범죄종류리스트[1:],column='범죄종류')
    최종리스트.append((년도,프레임))
    print(년도)
```

```python
데이터프레임리스트=[]
for 년, 데이터프레임 in 최종리스트 :
    데이터프레임=데이터프레임.drop('3')
    직군리스트=데이터프레임.ix['1'].values
    직군=[]
    for 상자 in 직군리스트 :
        if not pd.isnull(상자) :
            num=상자.find('(')
            이름=상자[:num]
        직군.append(이름) 
    직군[-1]='미상'
    
    직군리스트=직군
    count=0
    
    for 년도 in 데이터프레임 :
        데이터프레임[년도][0]=직군리스트[count]
        count=count+1
    
    for 년도 in 데이터프레임 :
        if pd.isnull(데이터프레임[년도][1]) :
            데이터프레임[년도][1]=데이터프레임[년도][0]+'(합)'
    
    콜론명=데이터프레임.ix['2'].values
    콜론명[0]='범죄종류'
    데이터프레임.columns = 콜론명
    데이터프레임 = 데이터프레임.drop('2')
    
    인덱스명=데이터프레임.index.values
    인덱스명[0]='직종'
    
    데이터프레임=데이터프레임.T
    데이터프레임리스트.append((년,데이터프레임))
```

```python
writer=pd.ExcelWriter(os.path.join(타겟경로,파일이름))
for 년도, 데이터프레임 in 데이터프레임리스트 :
    데이터프레임.to_excel(writer,sheet_name=년도)    
writer.save()
```

```python
import pandas as pd
import os
```

```python
포맷파일디렉토리="C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일"
파일이름="범죄발생 지역(1964-2013).xlsx"

데이터프레임=pd.read_excel(os.path.join(포맷파일디렉토리,파일이름),encoding='utf-8')
데이터프레임=데이터프레임.drop(' 범죄명칭.2',axis=1) #1열방향
데이터프레임=데이터프레임.T.drop(0,axis=1)
데이터프레임
```

```python
콜론=[]
for i in 데이터프레임.columns :
    이름=""
    for name in ['범죄명칭',' 범죄명칭',' 범죄명칭.1'] :
        이름 += 데이터프레임.ix[name][i] +'/'
    콜론.append(이름)

데이터프레임.columns = 콜론
데이터프레임=데이터프레임.reset_index(drop=True)
```

```python
데이터프레임=데이터프레임.drop([0,1,2])
```

```python
데이터프레임.columns = 데이터프레임.columns[1:].insert(0,'도시')
데이터프레임=데이터프레임.reset_index(drop=True)
데이터프레임
```

```python
데이터프레임=데이터프레임.dropna()
```

```python
new데이터프레임=데이터프레임
```

```python
인덱스 = new데이터프레임['도시'].values
new데이터프레임=new데이터프레임.drop('도시',axis=1)
```

```python
new데이터프레임=new데이터프레임.set_index(인덱스,drop=True)
```

```python
import pandas as pd
import os
```

```python
포맷안된파일='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타켓파일='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄자 교육정도[1972-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된파일,파일이름))
데이터프레임
```

```python
대분류리스트=데이터프레임['범죄명칭'].values.tolist()


word=''
for i, 범죄 in enumerate(대분류리스트) :
    if pd.notnull(범죄) :
        word=범죄
    대분류리스트[i]=word
대분류리스트=[범죄 for 범죄 in 대분류리스트 if 범죄 != '']
대분류리스트
```

```python
소분류리스트=데이터프레임['Unnamed: 1'].values.tolist()
word=''
for i,소분류 in enumerate(소분류리스트) :
    if pd.notnull(소분류) :
        word=소분류
    소분류리스트[i]=word
소분류리스트=소분류리스트[3:]
소분류리스트
```

```python
범죄이름리스트=[]
for index in 데이터프레임.index :
    Temp=''
    for colum in ['Unnamed: 2','Unnamed: 3'] :
        word=데이터프레임.ix[index][colum]
        if pd.notnull(word) :
            Temp+=word+','
    if Temp=='' :
        Temp=데이터프레임.ix[index]['Unnamed: 1']
    else : Temp=Temp[:-1]
    범죄이름리스트.append(Temp)
for i in range(3) :범죄이름리스트[i]=i

데이터프레임=데이터프레임.set_index([범죄이름리스트])
데이터프레임=데이터프레임.drop(['범죄명칭','Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
```

```python
데이터프레임=데이터프레임.drop(2)
```

```python
데이터프레임리스트={year:[] for year in range(1991,2014)}
년도=''
for column in 데이터프레임 :
    try : 년도=int(column)
    except : pass
    데이터프레임리스트[년도].append(데이터프레임[column])
    
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(데이터프레임,axis=1)
    데이터프레임.columns = 데이터프레임.ix[1].values.tolist()
    데이터프레임=데이터프레임.drop([0,1])
    데이터프레임.insert(0,value=대분류리스트,column='대분류')
    데이터프레임.insert(1,value=소분류리스트,column='소분류')
    데이터프레임=데이터프레임.T
    데이터프레임리스트[년도]=데이터프레임
데이터프레임리스트[1991]
```

```python
writer=pd.ExcelWriter(os.path.join(타켓파일,파일이름))
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임.to_excel(writer,str(년도))
writer.save()
```

```python
import os
import pandas as pd
from collections import defaultdict
```

```python
포맷안된디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟디렉토리='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄 발생일 특수사정[1964-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된디렉토리,파일이름))
데이터프레임=데이터프레임.drop([2,3]).reset_index(drop=True)
데이터프레임
```

```python
범죄명칭리스트=[]
for index in 데이터프레임.index :
    temp=''
    for column in ['Unnamed: 1','Unnamed: 2','Unnamed: 3'] :
        word=데이터프레임.ix[index][column]
        if not pd.isnull(word) :
            temp+=word+','
    범죄명칭리스트.append(temp[:-1])

for i in range(2) : 범죄명칭리스트[i]=i
데이터프레임=데이터프레임.set_index([범죄명칭리스트])
데이터프레임=데이터프레임.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
데이터프레임
```

```python
범죄범주리스트=[명칭 for 명칭 in 데이터프레임['범죄명칭'].tolist()]
word=''
for i,범죄범주 in enumerate(범죄범주리스트) :
    if pd.notnull(범죄범주) :
        word=범죄범주
    범죄범주리스트[i]=word

데이터프레임=데이터프레임.drop('범죄명칭',axis=1)
데이터프레임
```

```python
데이터프레임리스트={str(year):[] for year in range(1991,2014)}

year=''
for 콜론 in 데이터프레임 :
    for 년도 in range(1991,2014) :
        if 콜론 == str(년도) :
            year=년도
            break;
    데이터프레임리스트[str(year)].append(데이터프레임[콜론])
```

```python
for 년도,데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(데이터프레임,axis=1)
    데이터프레임=데이터프레임.drop(0)
    데이터프레임.columns=데이터프레임.ix[0].values
    데이터프레임=데이터프레임.drop(1)
    데이터프레임.insert(0,value=범죄범주리스트[2:],column='범죄범주')
    데이터프레임리스트[년도]=데이터프레임
```

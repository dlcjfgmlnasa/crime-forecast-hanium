```python
import pandas as pd
import os
```

```python
파일경로="C:/Users/chlee/Desktop/crime-forecast-hanium"
파일이름="2014년 - 범죄자특성 - 사회인구학적 특성 - 범죄자 교육정도(2014).xlsx"

데이터프레임=pd.read_excel(os.path.join(파일경로,파일이름),encoding='utf-8', sheetName="", index=False)
데이터프레임.drop(0)
데이터프레임=데이터프레임.T #행과 열의 위치를 바꾼다
데이터프레임=데이터프레임.drop(0,axis=1)
학력=데이터프레임[1]
데이터프레임.set_index(학력).drop(1,axis=1)
데이터프레임
```

```python
Temp=[]
for i in range(3) :
    Temp.append(데이터프레임.ix[i].T)
콜론데이터프레임=pd.concat(Temp,axis=1)

Temp=[]
for i in  콜론데이터프레임.index :
    콜론=''
    for columns in 콜론데이터프레임.columns :
        col = 콜론데이터프레임.ix[i][columns].strip()
        if  col == "" : continue
        콜론 += col + '/'
    Temp.append(콜론)
```

```python
데이터프레임.columns = Temp #콜론의 이름을 바꾼다.
데이터프레임 = 데이터프레임.drop(['범죄명칭',' 범죄명칭',' 범죄명칭.1'])
```

```python
데이터프레임 = 데이터프레임.reset_index(drop=True) #이렇게 하면 기존의 colum을 삭제한다. 아니면 column의 내용이 colum으로 들어간다.
데이터프레임
```

```python
데이터프레임 = 데이터프레임.dropna() #NAN의 내용을 삭제한다.
```

```python
파일이름 = '2014년 범죄자 교육정도.csv'
데이터프레임.to_csv(os.path.join(파일경로,파일이름),encoding='utf-8',index=False) #index를 Flase
```

```python
데이터프레임=pd.read_csv(os.path.join(파일경로,파일이름),encoding='utf-8',sep=',')
데이터프레임
```

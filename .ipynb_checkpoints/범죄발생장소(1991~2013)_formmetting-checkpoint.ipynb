```python
import pandas as pd
import os
```

```python
포맷안된파일='C:/Users/chlee/Desktop/crime-forecast-hanium/포맷안된파일'
타겟파일='C:/Users/chlee/Desktop/crime-forecast-hanium/데이터파일'
파일이름='범죄발생 장소[1972-2013].xlsx'
```

```python
데이터프레임=pd.read_excel(os.path.join(포맷안된파일,파일이름))
데이터프레임=데이터프레임.drop(2)
데이터프레임
```

```python
범죄범주=데이터프레임['범죄명칭'].values.tolist()
word=''
for i,범주 in enumerate(범죄범주) :
    if pd.notnull(범주) :
        word=범주
    범죄범주[i]=word
for i in range(2) : 범죄범주[i]=i
데이터프레임=데이터프레임.drop('범죄명칭',axis=1)
데이터프레임
```

```python
범죄명칭리스트=[]
for 인덱스 in 데이터프레임.index :
    Temp=''
    for 콜론 in ['Unnamed: 1','Unnamed: 2','Unnamed: 3'] :
        word=데이터프레임.ix[인덱스][콜론]
        if pd.notnull(word) :
            Temp+=word+','
    범죄명칭리스트.append(Temp[:-1])
for i in range(2) : 범죄명칭리스트[i]=i
데이터프레임=데이터프레임.drop(['Unnamed: 1','Unnamed: 2','Unnamed: 3'],axis=1)
데이터프레임=데이터프레임.set_index([범죄명칭리스트])
데이터프레임
```

```python
데이터프레임리스트={str(year):[] for year in range(1991,2014)}
Temp=''
for 년도 in 데이터프레임 :
    for year in range(1991,2014) :
        if 년도 == str(year) :
            Temp=년도
            break
    데이터프레임리스트[Temp].append(데이터프레임[년도])
```

```python
for 년도, 데이터프레임 in 데이터프레임리스트.items() :
    데이터프레임=pd.concat(데이터프레임,axis=1)
    데이터프레임.columns=데이터프레임.ix[1].values.tolist()
    데이터프레임=데이터프레임.drop([0,1])
    데이터프레임.insert(0,value=범죄범주[2:],column='범죄범주')
    데이터프레임=데이터프레임.T
    #데이터프레임=데이터프레임.fillna(0) #nan을 0으로 바꾼다.
    데이터프레임리스트[년도]=데이터프레임

데이터프레임리스트['1991']
```
